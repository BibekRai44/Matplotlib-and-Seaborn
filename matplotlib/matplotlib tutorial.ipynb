```python
from matplotlib import pyplot as plt
x=[4,6,3]
y=[6,7,2]
x1=[7,3,7]
y1=[2,7,8]
plt.scatter(x,y,label="first",linewidth=4)
plt.hist(x1,y1,histtype='bar',rwidth=0.8,label="second",linewidth=4)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[4,6,3]
y=[6,7,2]
x1=[7,3,7]
y1=[2,7,8]
plt.bar(x,y,label="first",linewidth=3)
plt.bar(x1,y1,label="second",linewidth=3)
plt.legend()
plt.title('Broke')
plt.show()
```

```python
from matplotlib import pyplot as plt
days=[1,2,3,4,5]
working=[2,5,2,5,7]
eating=[5,2,7,2,7]
running=[5,7,2,5,7]
sleeping=[8,3,5,2,7]
plt.plot([],[],color='m',label='working',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='running',linewidth=5)
plt.plot([],[],color='k',label='sleeping',linewidth=5)
plt.stackplot(days,working,eating,running,sleeping,colors=['m','c','r','k'])
plt.legend()
plt.title('Broke')
plt.show()
```

```python
from matplotlib import pyplot as plt
days=[1,2,3,4]
cols=['m','c','r','k']
activities=['working','eating','running','sleeping']
plt.pie(days,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')
plt.title('pieplot')
plt.show()
```
